# Broadcasting in Python
Deep learning 2022


Broadcasting is the term used to describe the implicit element-by-element behavior of operations. Could be multidimensional
arrays of the same shape, or a scalar and an array, or even two arrays of with different shapes, provided that the smaller array is “expandable” to the shape of the larger in such a way that th resulting broadcast is unambiguous. - Numpy documentation.

```python
import numpy as np
```

In the next matrix, I've shown the number of calories from carbohydrates, proteins, and fats in 100 grams of four different foods.

```python
A = np.array([[56,0.0,4.4,68.0],
             [1.2,104.0,52.0,8.0],
             [1.8,135.0,99.0,0.9]])
```

```python
print(A)
```

First, we're going to compute "cal" equals A dot sum. And axis equals 0 means to sum vertically.

```python
cal = A.sum(axis=0)
print(cal)
```

```python
print("Cálculo de la dimensión del row-vector 'cal': ",cal.shape) # La dimensión resultante es de una matriz de rango 1
print("Cálculo de la dimensión de la matriz 'A': ",A.shape)
```

And then we're going to compute percentage equals A/cal.reshape 1,4.

```python
#porcentaje = 100*A/cal.reshape(1,4)
porcentaje = 100*A/cal
print(porcentaje)
```

```python
print("Reshape del vector 'cal'...")
cal = cal.reshape(1,4)
print("Nueva dimensión del row-vector 'cal': ",cal.shape)
```

```python
#porcentaje = 100*A/cal.reshape(1,4)
porcentaje = 100*A/cal
print(porcentaje)
```

## A note on Python/Numpy vectors - broadcasting

```python
a = np.random.randn(5)
print (a)
print("\nDimensión del arreglo aleatorio 'a': ", a.shape, " - La dimensión es de una matriz de rango 1")
```

```python
print("Cálculo de la transpuesta de la matriz de rango 1 'a':\n",a.T)
print("\n¡No existe cambio en la estructura de la matriz inicial y la transpuesta!")
```

```python
print("Cálculo del producto vectorial de 'a' y su transpuesta: ", np.dot(a,a.T))
print("\nOjito: el resultado obtenido fue un número, no el producto entre vectores...")
```

```python
a = np.random.randn(5,1)
print ("Column-vector:\n",a,"\n\nDimensión del column-vector: ",a.shape)
print("\nTranspuesta del column-vector:\n",a.T)
```

```python
assert(a.shape == (5,1))
```

```python
print("Cálculo del producto vectorial de 'a' y su transpuesta:\n", np.dot(a,a.T))
```

```python
a = np.random.randn(1,5)
print ("Row-vector:\n",a,"\n\nDimensión del row-vector: ",a.shape)
print("\nTranspuesta del Row-vector:\n",a.T)
```
